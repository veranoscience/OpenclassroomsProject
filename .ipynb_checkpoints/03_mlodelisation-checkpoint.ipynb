## Réalisation d'un premier modéle de classification

```python
def eval_side(X, y, model, set_name="TEST", plot_cm=True):
    """ Fonction pour calculer et afficher les métriques clés + matrice de confusion pour un jeu (train/test)."""
    y_pred  = model.predict(X)
    y_proba = model.predict_proba(X)[:, 1] if hasattr(model, "predict_proba") else None

    acc  = accuracy_score(y, y_pred)
    prec = precision_score(y, y_pred, pos_label=1, zero_division=0)
    rec  = recall_score(y, y_pred,    pos_label=1, zero_division=0)
    f1   = f1_score(y, y_pred,        pos_label=1, zero_division=0)
    roc  = roc_auc_score(y, y_proba) if y_proba is not None else np.nan
    pr   = average_precision_score(y, y_proba) if y_proba is not None else np.nan

    print(f"\n--- {set_name} ---")
    print(f"Accuracy : {acc:.3f} | Précision : {prec:.3f} | Rappel : {rec:.3f} | F1 : {f1:.3f}")
    if y_proba is not None:
        print(f"ROC AUC : {roc:.3f} | PR AUC : {pr:.3f}")
    print("\nClassification report :\n", classification_report(y, y_pred, digits=3))

    if plot_cm:
        cm = confusion_matrix(y, y_pred, labels=[0, 1])
        disp = ConfusionMatrixDisplay(cm, display_labels=["Restés (0)", "Partis (1)"])
        fig, ax = plt.subplots(figsize=(4,4))
        disp.plot(ax=ax, cmap="Blues", values_format="d", colorbar=False)
        ax.set_title(f"Matrice de confusion – {set_name}")
        plt.tight_layout(); plt.show()

    return {"accuracy":acc, "precision":prec, "recall":rec, "f1":f1, "roc_auc":roc, "pr_auc":pr}
```

```python
def fit_and_evaluate(name, model, X_train, y_train, X_test, y_test):
    """Fonction pour entraîner le modèle sur TRAIN et évaluer sur TRAIN et TEST. Retourne les scores."""
    print(f"\n================= {name} =================")
    model.fit(X_train, y_train)
    scores_train = eval_side(X_train, y_train, model, set_name="TRAIN")
    scores_test  = eval_side(X_test,  y_test,  model, set_name="TEST")
    return {"train": scores_train, "test": scores_test}
```

```python
models = [
    ("Dummy", DummyClassifier(strategy="most_frequent", random_state=42)),
    ("Logistic Regression (balanced)",
     LogisticRegression(solver="liblinear", class_weight="balanced", random_state=42)),
    ("RandomForest (balanced)",
     RandomForestClassifier(n_estimators=300, class_weight="balanced",
                            random_state=42))
]
```

```python
# --- Boucle d’évaluation ---
all_scores = {}
for name, mdl in models:
    all_scores[name] = fit_and_evaluate(name, mdl, X_train, y_train, X_test, y_test)
```

```python
# Fonction pour evaluer tous les modèles et stocker dans 2 DataFrames (TRAIN / TEST)
rows_train, rows_test = [], []

for name, mdl in models:
    # fit sur TRAIN
    mdl.fit(X_train, y_train)

    # métriques TRAIN (sans afficher de matrice)
    m_tr = eval_side(X_train, y_train, mdl, set_name="TRAIN", plot_cm=False)
    m_tr.update({"model": name, "set": "TRAIN"})
    rows_train.append(m_tr)

    # métriques TEST (sans afficher de matrice)
    m_te = eval_side(X_test, y_test, mdl, set_name="TEST", plot_cm=False)
    m_te.update({"model": name, "set": "TEST"})
    rows_test.append(m_te)
```

```python
# Construire les DataFrames nettoyés et ordonnés
cols = ["model","set","accuracy","precision","recall","f1","roc_auc","pr_auc"]
df_metrics_train = pd.DataFrame(rows_train)[cols].round(3)
df_metrics_test  = pd.DataFrame(rows_test)[cols].round(3)
```

```python
display(" METRICS – TRAIN ")
display(df_metrics_train)
display("METRICS – TEST")
display(df_metrics_test)
```

Examinons de plus près l'utilisation de l'algorithme de régression logistique. 
J'utiliserai une validation croisée à 10 plis pour entraîner notre modèle de régression logistique et estimer son score AUC.

### Logistic Regression

```python
logit_bal = LogisticRegression(
    solver="liblinear",            
    class_weight="balanced",
    max_iter=1000,
    random_state=42
)
```

```python
# Préprocesseur : scaler uniquement les numériques
preproc = ColumnTransformer(
    transformers=[("num", StandardScaler(), num_cols)],
    remainder="passthrough"
)
```

```python
def cv_scores_summary(model, X_train, y_train, n_splits=10, random_state=42):
    
    """CV sur X_train/y_train avec stratification + scaling dans la Pipeline."""
    
    cv   = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=random_state)

    pipe = Pipeline(steps=[("prep", preproc), ("clf", model)])

    s_roc = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="roc_auc",            n_jobs=-1)
    s_pr  = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="average_precision",  n_jobs=-1)
    s_acc = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="accuracy",           n_jobs=-1)

    df_folds = pd.DataFrame({
        "fold": np.arange(1, n_splits+1),
        "roc_auc": s_roc,
        "pr_auc":  s_pr,
        "accuracy": s_acc
    })
    
    summary = (
        df_folds[["roc_auc","pr_auc","accuracy"]]
        .agg(["mean","std"]).T.reset_index()
        .rename(columns={"index":"metric"})
    )
    return df_folds, summary
```

```python
# Logit équilibré
df_logit_folds, sum_logit = cv_scores_summary(logit_bal, X_train, y_train, n_splits=10)

display("Logit – CV (moyenne et STD)", sum_logit.round(3))
```

**À retenir comme métrique principale : PR AUC**

**PR AUC (average precision)** : métrique la plus pertinente quand la classe “partis” est minoritaire.

**ROC AUC**: qualité de tri des scores (utile mais moins sensible à l’imbalance que PR AUC).

**Accuracy** : à prendre avec beaucoup de recul (peut être trompeuse en cas de déséquilibre).

### Random Forest

```python
rf_bal = RandomForestClassifier(
    n_estimators=100,
    class_weight="balanced",
    max_depth=15,
    min_samples_leaf=3,
    max_features="sqrt",
    n_jobs=-1,
    random_state=42
)
```

```python
# Random Forest équilibré
df_rf_folds, sum_rf = cv_scores_summary(rf_bal, X_train, y_train, n_splits=10)
display("RF – CV (moyenne et std):\n", sum_rf.round(3))
```

### Choisir un seuil via la courbe Précision–Rappel (sur TRAIN en CV)

On calcule des prédictions out-of-fold (OOF) puis on choisit un seuil F2 

On utilise `cross_val_predict(..., method="predict_proba")` en CV stratifiée sur le train pour calculer la courbe PR et choisir un seuil.

```python
# Pipeline final = prétraitement + sur la modèle (logistique balanced)
pipe = Pipeline([("prep", preproc), ("clf", rf_bal)])
```

```python
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
```

```python
# Probabilités OOF sur TRAIN (classe 1 = "partis")
proba_oof = cross_val_predict(
    pipe, X_train, y_train.squeeze(), cv=cv, n_jobs=-1, method="predict_proba"
)[:, 1]
```

```python
# Courbe PR
prec, rec, thr = precision_recall_curve(y_train, proba_oof)
ap = average_precision_score(y_train, proba_oof)
print(f"Average Precision (PR AUC) OOF = {ap:.3f}")
```

```python
# Visualisations Courve PR
plt.figure(figsize=(6,4))
plt.plot(rec, prec)[[’
plt.xlabel("Rappel")
plt.ylabel("Précision")
plt.title("Courbe Précision–Rappel (OOF)")
plt.grid(True); plt.tight_layout(); plt.show()
```

```python
# F1 pour chaque seuil
f1 = 2 * prec * rec / (prec + rec + 1e-12)
best_idx = f1.argmax()
best_thr = thr[best_idx]
print(f"Seuil optimal (max F1) = {best_thr:.3f} | P={prec[best_idx]:.3f} R={rec[best_idx]:.3f} F1={f1[best_idx]:.3f}")
```

```python
plt.figure(figsize=(6,4))
plt.plot(thr, f1[:-1])
plt.xlabel("Seuil")
plt.ylabel("F1")
plt.title("F1 en fonction du seuil (OOF)")
plt.grid(True); plt.tight_layout(); plt.show()
```

Seuil RF “F1-optimum” ≈ 0.35–0.40. À ce seuil, compromis correct; en baissant le seuil je gagne du rappel, en l’augmentant je gagne en précision.

Le RF montre encore des signes d’overfit

**Choix d'optimisation (métiers -> seuil)** Notre priorité est de ne pas rater des démissions, en acceptant une baisse de précision. Pour ça, on privilégie un seuil plus bas pour augmenter le rappel (capturer un maximum de futurs départs)

```python
prec, rec, thr = precision_recall_curve(y_train, proba_oof)
```

```python
# Fbeta score

f2 = (1+2**2) * (prec*rec) / (2**2*prec + rec + 1e-12)   # Fβ formule
best_f2_idx = np.nanargmax(f2)
thr_f2 = thr[best_f2_idx]
p_f2, r_f2 = prec[best_f2_idx], rec[best_f2_idx]
```

```python
print(f"Seuil F2 : {thr_f2:.3f} | P={p_f2:.3f} R={r_f2:.3f}")
```

```python
# Fitte sur tout TRAIN et évaluation sur TEST
pipe.fit(X_train, y_train)
proba_test = pipe.predict_proba(X_test)[:, 1]
y_pred_test = (proba_test >= thr_f2).astype(int)
```

```python
pr_auc  = average_precision_score(y_test, proba_test)
roc_auc = roc_auc_score(y_test, proba_test)
```

```python
print("\n=== Évaluation TEST (seuil F2) ===")
print(classification_report(y_test, y_pred_test, digits=3))
print(f"PR AUC (test) : {pr_auc:.3f}")
print(f"ROC AUC (test): {roc_auc:.3f}")
```

Classe 1 (partis) → Précision 0.34, Rappel 0.745, F1 0.467.

C’est cohérent avec un seuil optimisé pour F2 : on privilégie le rappel (on capte ~75% des départs), au prix de faux positifs.

Global → Accuracy 0.728 (pas très pertinente avec classes déséquilibrées).

TN (vrais négatifs) = 179 : restés correctement prédit “restés”.
